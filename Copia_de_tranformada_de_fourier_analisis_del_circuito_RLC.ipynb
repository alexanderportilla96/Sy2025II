**Valores de los Componentes:**
* **Resistencia (R):** $1 \text{ k}\Omega = 1000 \text{ }\Omega$
* **Inductancia (L):** $180 \text{ mH} = 0.18 \text{ H}$
* **Capacitancia (C):** $120 \text{ }\mu\text{F} = 0.00012 \text{ F}$

---

### 1. Modelo en el Dominio de Laplace (Dominio 's')

Primero, representamos el circuito en el dominio 's' asumiendo que es un circuito RLC en serie.


[Image of a series RLC circuit diagram]


Las impedancias de cada componente son:
* **Resistor:** $Z_R = R = 1000$
* **Inductor:** $Z_L = sL = 0.18s$
* **Capacitor:** $Z_C = \frac{1}{sC} = \frac{1}{0.00012s}$

La impedancia total del circuito $Z(s)$ es la suma de todas las impedancias en serie:
$$Z(s) = Z_R + Z_L + Z_C = 1000 + 0.18s + \frac{1}{0.00012s}$$

---

### 2. Función de Transferencia H(s)

Vamos a definir la función de transferencia para la corriente $I(s)$ en respuesta a un voltaje de entrada $V_{in}(s)$.

$$H(s) = \frac{I(s)}{V_{in}(s)}$$

Usando la ley de Ohm en el dominio 's' ($V = I \cdot Z$):
$$I(s) = \frac{V_{in}(s)}{Z(s)}$$
$$H(s) = \frac{1}{Z(s)} = \frac{1}{1000 + 0.18s + \frac{1}{0.00012s}}$$

Para simplificar, multiplicamos el numerador y el denominador por $s$:
$$H(s) = \frac{s}{1000s + 0.18s^2 + \frac{1}{0.00012}}$$

Calculamos el término $\frac{1}{C}$:
$$\frac{1}{0.00012} \approx 8333.33$$

Reordenando en la forma polinómica estándar (orden descendente de $s$):
$$H(s) = \frac{s}{0.18s^2 + 1000s + 8333.33}$$

---

### 3. Análisis de la Respuesta (Cálculo de Polos)

El comportamiento del sistema (si es sobreamortiguado, subamortiguado o críticamente amortiguado) depende de las **raíces del denominador** (los polos del sistema).

Buscamos los valores de $s$ que hacen cero al denominador:
$$0.18s^2 + 1000s + 8333.33 = 0$$

Usamos la fórmula cuadrática $s = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$
* $a = 0.18$
* $b = 1000$
* $c \approx 8333.33$

Calculamos el discriminante ($\Delta = b^2 - 4ac$):
* $b^2 = (1000)^2 = 1,000,000$
* $4ac = 4 \cdot 0.18 \cdot 8333.33 = 0.72 \cdot 8333.33 \approx 6000$

$$\Delta = 1,000,000 - 6000 = 994,000$$

**Conclusión del Amortiguamiento:**
Dado que el discriminante $\Delta > 0$ (es un número real positivo), el sistema tiene **dos polos reales y distintos**. Esto significa que el circuito es **SOBREAMORTIGUADO (Overdamped)**.

---

### 4. Polos del Sistema

Ahora calculemos los valores exactos de los polos:
$$\sqrt{\Delta} = \sqrt{994000} \approx 996.995$$

* **Polo 1 ($s_1$):**
    $$s_1 = \frac{-1000 + 996.995}{2 \cdot 0.18} = \frac{-3.005}{0.36} \approx -8.347$$

* **Polo 2 ($s_2$):**
    $$s_2 = \frac{-1000 - 996.995}{2 \cdot 0.18} = \frac{-1996.995}{0.36} \approx -5547.2$$

Los polos están en $s_1 \approx -8.35$ y $s_2 \approx -5547.2$.

---

### 5. Interpretación de la Respuesta

* **Tipo de Respuesta:** Sobreamortiguada.
* **Respuesta al Escalón:** Si aplicas un voltaje de entrada constante (como una batería de 1V, $V_{in}(s) = \frac{1}{s}$), la corriente $i(t)$ **no oscilará**.
* **Comportamiento:** La corriente subirá a un valor pico y luego decaerá lentamente hacia cero, siguiendo una curva compuesta por dos exponenciales decrecientes ($e^{-8.35t}$ y $e^{-5547.2t}$). El polo $s_1 = -8.35$ (el más cercano a cero) es el **polo dominante** y dictará la velocidad de la respuesta, ya que el término $e^{-5547.2t}$ desaparecerá casi instantáneamente.

En resumen, la alta resistencia (1 k$\Omega$) domina sobre la inductancia y la capacitancia, "frenando" el sistema e impidiendo cualquier oscilación.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Definiciones Simbólicas y Constantes ---

# Definir las variables simbólicas
t, s = sp.symbols('t s')

# Definir los valores de los componentes
R = 1000      # 1 kOhm
L = 0.18      # 180 mH
C = 0.000120  # 120 uF

print("Análisis de Circuito RLC Serie")
print(f"R = {R} Ohms, L = {L} H, C = {C} F")
print("="*40)


# --- 2. Cálculo de H(s) (Función de Transferencia) ---
# H(s) = I(s) / V(s)

# Impedancias en el dominio 's'
Z_R = R
Z_L = s * L
Z_C = 1 / (s * C)

# Impedancia total Z(s)
Z_s = Z_R + Z_L + Z_C

# Función de Transferencia H(s) = 1 / Z(s)
H_s = 1 / Z_s

# Simplificar H(s) para verla en formato polinómico
H_s_simplified = sp.simplify(H_s)

print("\nFunción de Transferencia H(s):")
sp.pprint(H_s_simplified)
print("="*40)


# --- 3. Cálculo de h(t) (Respuesta al Impulso) ---
# h(t) = L_inv{H(s)}

print("\nCalculando h(t) = L_inv{H(s)}...")
h_t = sp.inverse_laplace_transform(H_s_simplified, s, t)

print("Respuesta al Impulso h(t):")
sp.pprint(h_t)
print("="*40)


# --- 4. Cálculo de y(t) (Respuesta al Escalón) ---
# y(t) = i(t) para una entrada de escalón unitario V_in(t) = 1u(t)
# V_in(s) = 1/s

V_in_s = 1/s

# Y(s) = H(s) * V_in(s)
Y_s = H_s_simplified * V_in_s

# Simplificar Y(s) (el 's' del numerador de H(s) y el 's' de V(s) se cancelan)
Y_s_simplified = sp.simplify(Y_s)

print(f"\nRespuesta en 's' al escalón Y(s) = H(s) * (1/s):")
sp.pprint(Y_s_simplified)

print("\nCalculando y(t) = L_inv{Y(s)}...")
y_t = sp.inverse_laplace_transform(Y_s_simplified, s, t)

print("Respuesta al Escalón y(t) [Corriente i(t)]:")
sp.pprint(y_t)
print("="*40)


# --- 5. Graficación de las Respuestas ---

print("Generando gráficos de h(t) y y(t)...")

# Convertir las expresiones simbólicas a funciones numéricas
# Usamos .subs(sp.Heaviside(t), 1) porque estamos ploteando para t > 0
h_t_numeric = sp.lambdify(t, h_t.subs(sp.Heaviside(t), 1), 'numpy')
y_t_numeric = sp.lambdify(t, y_t.subs(sp.Heaviside(t), 1), 'numpy')

# Creamos un vector de tiempo
# El polo dominante es ~ -8.347, la constante de tiempo es 1/8.347 ≈ 0.12s
# Graficamos por 5 constantes de tiempo (5 * 0.12s = 0.6s)
t_vals = np.linspace(0, 0.6, 500)

# Calcular los valores de corriente
h_vals = h_t_numeric(t_vals)
y_vals = y_t_numeric(t_vals)

# Crear los gráficos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

# Gráfico para h(t)
ax1.plot(t_vals, h_vals, label='h(t) - Respuesta al Impulso', color='blue')
ax1.set_title(f'h(t) - Respuesta al Impulso\n R={R}Ω, L={L}H, C={C}F')
ax1.set_ylabel('Corriente (A)')
ax1.grid(True)
ax1.legend()

# Gráfico para y(t)
ax2.plot(t_vals, y_vals, label='y(t) - Respuesta al Escalón', color='red')
ax2.set_title(f'y(t) - Respuesta al Escalón (Corriente i(t))\n R={R}Ω, L={L}H, C={C}F')
ax2.set_xlabel('Tiempo (s)')
ax2.set_ylabel('Corriente (A)')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()

print("Análisis completo.")
```
