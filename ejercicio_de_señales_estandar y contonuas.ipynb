```python
import sympy as sym

# Define symbols
t, omega = sym.symbols('t omega')

# Define a cosine signal
cosine_wave = sym.cos(omega * t).subs(omega, 2)

# Generate the full-wave rectified signal
rectified_wave = cosine_wave * sym.sign(cosine_wave)

# Plot the original and rectified signals
p1 = sym.plot(cosine_wave, (t, 0, 4*sym.pi), label='Original', show=False)
p2 = sym.plot(rectified_wave, (t, 0, 4*sym.pi), label='Rectified', show=False, line_color='red')

p1.extend(p2)
p1.legend = True
p1.ylabel = r'$x(t)$'
p1.show()
```
